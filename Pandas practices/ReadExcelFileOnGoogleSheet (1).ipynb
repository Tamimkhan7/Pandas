```python
# how to rename a column and concate dataframe

df1 = pd.DataFrame({'AGE' : [16, 20], 'INCOME' : [3000, 4000]})
df2 = pd.DataFrame({'AGE' : [25, 30], 'INCOME' : [7000, 3000]})
df1
```

```python
df2
```

```python
df1 = df1.rename(columns={'AGE' : 'age', 'INCOME' : 'income'})
df1
```

```python
df2 = df2.rename(columns={'AGE' : 'age', 'INCOME' : 'income'})
df2
```

```python
# if i have declear axis is 1 that is show column shape, if i have declear axis=0 that is show row shape, it is default is 0
res = pd.concat([df1,df2], axis=1)
res
```
