```python
import pandas as pd

url = 'https://raw.githubusercontent.com/rashakil-ds/5-Minutes-to-Pandas/main/Datasets/Screen%20Time%20Data.csv'
df = pd.read_csv(url)
df.head()
```

```python
# akta column ar max value kivabe ber kore

df['Total Screen Time '].max()
```

```python
# akta column ar min value kivabe ber kore 

df['Total Screen Time '].min()
```

```python
# akta column ar kivabe average value ber kore

df['Total Screen Time '].mean()
```

```python
# std mane holo :- standard deviation mane holo ai gula average theke kotota kace or dure ace value gula
df['Total Screen Time '].std()
```

```python
# maximum kon value ta aita ber korar jonno amra mode use kori

df['Entertainment'].mode()
```

```python
df['Entertainment'].value_counts()
```

```python
# 25% modde max value
df['Total Screen Time '].quantile(0.25)
```

```python
# 75% modde max value
df['Total Screen Time '].quantile(0.75)
```

```python
# 100% modde max value
df['Total Screen Time '].quantile(1)
```

```python
# 0% modde max value hoye jabe min value
df['Total Screen Time '].quantile(0)
```

```python
df.describe()
```

```python
df['Social Networking'].sort_values()
```

```python
df['Social Networking'].sort_values(ascending= False)
```

```python
# nth largest column jodi ami dekhte cai tahoole akta specific column base kore 

df.nlargest(10, 'Social Networking')
```

```python
# nth smallest column jodi ami dekhte cai tahoole akta specific column base kore 

df.nsmallest(10, 'Social Networking')
```
