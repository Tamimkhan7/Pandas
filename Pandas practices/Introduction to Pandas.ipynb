```python
import numpy as np
import pandas as pd
```

```python
data = np.random.rand(10,3)
for col in range(data.shape[1]):
    null_indices = np.random.choice(10,3, replace=False)
    data[null_indices, col] = np.nan
df = pd.DataFrame(data, columns = ['Column_1', 'Column_2', 'Column_3'])
df 
```

```python
# check each column which one is null, is each matrix index is null that's show the true otherwise show the false
pd.isna(df)
```

```python
# best easyst way to check each column and row, which one is null or store value 

df.isnull()
```

```python
df.isnull().sum()
```

```python
# default 'how' thake any, mane holo, kono akta row te jodi null value thake oi row ta show korbe na, 
df.dropna()
```

```python
# akta row ar protita value jodi null hoy tahole oi row ta ar show korbe na
df.dropna(how='all')
```

```python
# thresh mane holo amar oi row te 'thresh=2' mane thresh soman ba besi notnull hole oi row ta show korbe. 
df.dropna(thresh=2)
```

```python
df.dropna(thresh=4)
```
